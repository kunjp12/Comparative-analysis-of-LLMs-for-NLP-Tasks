```python
# Function to compare predicted and output sentiments
def calculate_accuracy(predicted, output):
    # Convert strings of lists to actual lists
    predicted_sentiments = eval(predicted)
    output_sentiments = eval(output)

    # Check if any of the predicted sentiments are in the output sentiments
    match = any(sentiment in output_sentiments for sentiment in predicted_sentiments)
    return match

# Apply the function to each row
df['match'] = df.apply(lambda row: calculate_accuracy(row['sentiment_predicted'], row['sentiment_output']), axis=1)

# Calculate the overall accuracy
accuracy = df['match'].mean()

accuracy_percentage = accuracy * 100  # Convert to percentage
accuracy_percentage
```

```python
# Define functions to create the new columns as specified

def exact_match(predicted, output):
    # Convert string representations of lists to actual lists
    predicted_list = eval(predicted)
    output_list = eval(output)
    # Check if both lists are equal
    return 1 if predicted_list == output_list else 0

def any_match(predicted, output):
    # Convert string representations of lists to actual lists
    predicted_list = eval(predicted)
    output_list = eval(output)
    # Check if any element of the predicted list is in the output list
    return 1 if any(sentiment in predicted_list for sentiment in output_list) else 0

def percent_match(predicted, output):
    # Convert string representations of lists to actual lists
    predicted_list = eval(predicted)
    output_list = eval(output)
    # Calculate the percentage of predicted sentiments that are in the output list
    match_count = sum(sentiment in output_list for sentiment in predicted_list)
    return (match_count / len(predicted_list)) * 100 if predicted_list else 0

# Apply these functions to the dataframe
df['exact_match'] = df.apply(lambda row: exact_match(row['sentiment_predicted'], row['sentiment_output']), axis=1)
df['any_match'] = df.apply(lambda row: any_match(row['sentiment_predicted'], row['sentiment_output']), axis=1)
df['percent_match'] = df.apply(lambda row: percent_match(row['sentiment_predicted'], row['sentiment_output']), axis=1)

# Display the first few rows of the dataframe to confirm the new columns
df.head()
```

```python
import matplotlib.pyplot as plt

# Calculate the counts of each match type
exact_match_count = df['exact_match'].sum()
any_match_count = df['any_match'].sum()
total_count = len(df)

# Percentages for each type
exact_match_percentage = (exact_match_count / total_count) * 100
any_match_percentage = (any_match_count / total_count) * 100
no_match_percentage = 100 - any_match_percentage

# Labels for the slices
labels = 'Exact Match', 'Any Match', 'No Match'
sizes = [exact_match_percentage, any_match_percentage, no_match_percentage]
colors = ['#ff9999','#66b3ff','#99ff99']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.tight_layout()
plt.title('Sentiment Prediction Accuracy Visualization')
plt.show()
```

```python
# For x_train
text_file = '/content/emb_results.txt'  # Replace with your file path
with open(text_file, 'r') as file:
    data = file.read()

# Splitting text by 'Index {number}' and parsing to list of numpy arrays
segments = data.split('Index ')[1:]  # Skip the first split as it's empty
x_train = []

for segment in segments:
    numbers_str = segment.split(': ')[1]  # Get the part after the colon
    # Remove the square brackets and split the string
    number_strings = numbers_str.strip('[]\n').split(',')

    # Convert each string in the list to an integer
    number_list = [float(num) for num in number_strings]
    # numbers_str = numbers_str[1:-2]
    # number_list = [float(num.strip()) for num in numbers_str.split(',')]
    x_train.append(np.array(number_list))

x_train = np.array(x_train)  # Convert list of arrays to a 2D numpy array
```

```python
y_train = df['Answer'].values
y_train = y_train.astype('int')
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
X_train, X_test, Y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

# Create a Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
rf_classifier.fit(X_train, Y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Display classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
# Create a Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Define the parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create GridSearchCV object
grid_search = GridSearchCV(rf_classifier, param_grid, cv=3, scoring='accuracy')

grid_search.fit(X_train, Y_train)

# Print the best parameters
print("Best Parameters:", grid_search.best_params_)

# Get the best model
best_model = grid_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Display classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import multilabel_confusion_matrix, classification_report

# Assuming y_test and y_pred are already defined

# Convert multilabel predictions to binary
y_pred_binary = np.round(y_pred)

# Create multilabel confusion matrix
conf_matrix_multilabel = multilabel_confusion_matrix(y_test, y_pred_binary)

# Display classification report
print('Classification Report:')
print(classification_report(y_test, y_pred_binary))

# Calculate the number of rows and columns for subplots
num_labels = len(conf_matrix_multilabel)
num_cols = min(2, num_labels)  # Display at most 2 columns
num_rows = (num_labels + num_cols - 1) // num_cols

# Plot confusion matrix for each label
labels = range(num_labels)
plt.figure(figsize=(12, 8))
for i in labels:
    plt.subplot(num_rows, num_cols, i + 1)
    sns.heatmap(conf_matrix_multilabel[i], annot=True, fmt='d', cmap='Blues',
                xticklabels=['Not ' + str(labels[i]), str(labels[i])],
                yticklabels=['Not ' + str(labels[i]), str(labels[i])])
    plt.title(f'Confusion Matrix for Label {labels[i]}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')

plt.tight_layout()
plt.show()
```
